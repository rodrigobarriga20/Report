0. Imports

```python
import pandas as pd
```

1. DrugBank results

```python
# 3RQ3

df_3RQ3 = pd.read_csv("3RQ3_DrugBank_Results.csv")
df_3RQ3
```

```python
# 7VYT

df_7VYT = pd.read_csv("7VYT_DrugBank_Results.csv")
df_7VYT
```

```python
# Model

df_Model = pd.read_csv("Model_DrugBankResults.csv")
df_Model
```

2. DrugBank database

```python
# DrugBank (DB)

df_DB = pd.read_csv("DrugBank_14_03_2024.csv")
df_DB
```

```python
# Step 1: Aggregate all compounds and create df_all_docking
# Combine all unique compounds by "Title"
df_all_docking = pd.concat([df_3RQ3[['Title']], df_Model[['Title']], df_7VYT[['Title']]]).drop_duplicates().reset_index(drop=True)
df_all_docking.rename(columns={'Title': 'Database_ID'}, inplace=True)

# Add "Model", "3RQ3", and "7VYT" columns with "X" if the compound exists in the respective dataframe
df_all_docking['Model'] = df_all_docking['Database_ID'].isin(df_Model['Title']).map({True: 'X', False: ''})
df_all_docking['3RQ3'] = df_all_docking['Database_ID'].isin(df_3RQ3['Title']).map({True: 'X', False: ''})
df_all_docking['7VYT'] = df_all_docking['Database_ID'].isin(df_7VYT['Title']).map({True: 'X', False: ''})

# Step 2: Add new columns for relative energy, matched ligand sites, align score, and phase screen score
# Initialize columns for Model, 3RQ3, and 7VYT
for model in ['Model', '3RQ3', '7VYT']:
    df_all_docking[f'Relative energy - {model}'] = 'NA'
    df_all_docking[f'Matched Ligand Sites - {model}'] = 'NA'
    df_all_docking[f'Align Score - {model}'] = 'NA'
    df_all_docking[f'Phase Screen Score - {model}'] = 'NA'

# Define a function to update values from the respective dataframe
def update_values(df, source_df, source_name):
    for index, row in df.iterrows():
        match = source_df[source_df['Title'] == row['Database_ID']]
        if not match.empty:
            df.at[index, f'Relative energy - {source_name}'] = match['f3d_relative_energy'].values[0]
            df.at[index, f'Matched Ligand Sites - {source_name}'] = match['Matched Ligand Sites'].values[0]
            df.at[index, f'Align Score - {source_name}'] = match['Align Score'].values[0]
            df.at[index, f'Phase Screen Score - {source_name}'] = match['PhaseScreenScore'].values[0]

# Update df_all_docking with values from df_Model, df_3RQ3, and df_7VYT
update_values(df_all_docking, df_Model, 'Model')
update_values(df_all_docking, df_3RQ3, '3RQ3')
update_values(df_all_docking, df_7VYT, '7VYT')

# Step 3: Add columns from df_DB (DrugBank)
df_all_docking = pd.merge(df_all_docking, df_DB, left_on='Database_ID', right_on='DATABASE_ID', how='left')

# Drop duplicates and irrelevant columns
df_all_docking.drop(columns=['DATABASE_ID'], inplace=True)

# Sort by the largest value found in any of the three "Phase Screen Score" columns
df_all_docking.sort_values(by=['Phase Screen Score - Model', 'Phase Screen Score - 3RQ3', 'Phase Screen Score - 7VYT'], ascending=False, inplace=True)

# Save df_all_docking to CSV
df_all_docking.to_csv('All_Compounds_Pharmacophore.csv', index=False)

# Step 4: Create df_mostrelevantcompounds_pharmacophore
# Filter compounds that appear at least twice in different models
df_mostrelevantcompounds_pharmacophore = df_all_docking[(df_all_docking[['Model', '3RQ3', '7VYT']] == 'X').sum(axis=1) >= 2].copy()

# Count the number of "X"s in each row and add as a new column to prioritize sorting
df_mostrelevantcompounds_pharmacophore['X_count'] = (df_mostrelevantcompounds_pharmacophore[['Model', '3RQ3', '7VYT']] == 'X').sum(axis=1)

# Sort by "X_count" and then by highest Phase Screen Score across the three models
df_mostrelevantcompounds_pharmacophore.sort_values(by=['X_count', 'Phase Screen Score - Model', 'Phase Screen Score - 3RQ3', 'Phase Screen Score - 7VYT'], ascending=[False, False, False, False], inplace=True)

# Drop duplicates
df_mostrelevantcompounds_pharmacophore.drop_duplicates(subset=['Database_ID'], inplace=True)

# Save df_mostrelevantcompounds_pharmacophore to CSV
df_mostrelevantcompounds_pharmacophore.to_csv('Relevant_Compounds_Pharmacophore.csv', index=False)
```
