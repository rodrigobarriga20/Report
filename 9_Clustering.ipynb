```python
# Upload information and add Morgans to our dataset

dataset = pd.read_csv("Database_curated_WithMorgan.csv", sep = ";")
morgans = np.load("Morgan_Fingerprints.npy")

# Add Morgan Fingerprints to my Dataset dataframe

dataset["Morgan Fingerprints"] = morgans.tolist()
```

```python
# Each Morgan Fingerprint will be considered as one column and a new DataFrame with all Morgans will be made

columns = [f'Morgan{i}' for i in range (1,1025)]
Morgans = pd.DataFrame(morgans, columns=columns)
```

```python
# Normalization of Morgans

matrix = []
for index in Morgans.values:
    line = []
    for final_column in Morgans.values:
        column = np.logical_and(index, final_column).sum()/np.logical_or(index, final_column).sum()
        line.append(column)
    matrix.append(line)

matrix = np.array(matrix)
```

```python
# Clustering

clustering = AgglomerativeClustering(n_clusters=None, metric="precomputed", distance_threshold=0.7, linkage="complete")
clustering.fit(1-matrix)
cluster_id = clustering.labels_
```

```python
# With a cut-off of 70 % I can observe 3 clusters

cluster_id
```

```python
dataset["Cluster ID"] = cluster_id
dataset
```

```python
# Finding the maximum common structure of each cluster

mcs_list = []
no_cluster_count=0
for c in np.unique(cluster_id):
    mols_cluster = dataset[cluster_id==c]
    if len(mols_cluster)==1: no_cluster_count+=1;continue
    mcs = rdFMCS.FindMCS([Chem.MolFromSmiles(x) for x in mols_cluster.SMILES], completeRingsOnly=False,bondCompare=rdFMCS.BondCompare.CompareOrderExact,
                     ringMatchesRingOnly=True, matchChiralTag=False, matchValences=False,
                     maximizeBonds=False, timeout=1, verbose=True)
    mcs_list.append([c,len(mols_cluster),mcs.numAtoms,mcs.smartsString, Chem.MolFromSmiles(mols_cluster.SMILES.values[0])])
    
mcs_list = np.array(mcs_list)
```

```python
# Drawing of clusters
Drawing_Clusters = Draw.MolsToGridImage([x if x!="" else None for x in mcs_list[:,-1]], 
                     molsPerRow=5, subImgSize=(300, 200), highlightAtomLists=[x[4].GetSubstructMatch(Chem.MolFromSmarts(x[3])) for x in mcs_list],
                    legends=[f'cluster {x[0]}' for x in mcs_list], useSVG=True, returnPNG=False)

# Convert the SVG object to a string
svg_content = Drawing_Clusters.data

# Write the SVG content to a file
with open("Clusters.svg", "w") as f:
    f.write(svg_content)
```
